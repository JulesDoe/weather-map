```python
import pandas as pd
import requests

# Load data

# csv = pd.read_csv("data/biomass-all-story-urls-20210422214554.csv")
csv = pd.read_csv("data/carbon-neutrality-or-neu-all-story-urls-20210603205401.csv")
# csv = pd.read_csv("data/biomasse-or-bio-economie-all-story-urls-20210608163833.csv")

# Save JSON files

ids = csv['stories_id'].tolist()
len(ids)

for index, id in enumerate(ids):

    id = str(id)
    # filename = 'data/biomasse/' + id +'.json'
    filename = 'data/carbon_neutrality/' + id +'.json'

    try:
        f = open(filename)
        f.close()

    except:
        if (index % 10 == 0): print(index, id)
        r = requests.get('https://api.mediacloud.org/api/v2/stories_public/single/' + id + '?key=b2ef1a99a8fdbb84afafe742fd437c0942703072bb98242ba0eed9b8411e1735')
        open(filename, 'wb').write(r.content)
```
